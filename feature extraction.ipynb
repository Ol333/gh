```python
data1 = data[:656607]
data2 = data[656607:1360532]
data3 = data[1360532:2084747]
data4 = data[2084747:2788571]
data5 = data[2788571:]
```

```python
extraction_settings = ComprehensiveFCParameters()

X1 = extract_features(data1, column_id='id', column_sort='time',
                     default_fc_parameters=extraction_settings,
                     # we impute = remove all NaN features automatically
                     impute_function=impute)
```

```python
extraction_settings = ComprehensiveFCParameters()

X2 = extract_features(data2, column_id='id', column_sort='time',
                     default_fc_parameters=extraction_settings,
                     # we impute = remove all NaN features automatically
                     impute_function=impute)
```

```python
extraction_settings = ComprehensiveFCParameters()

X3 = extract_features(data3, column_id='id', column_sort='time',
                     default_fc_parameters=extraction_settings,
                     # we impute = remove all NaN features automatically
                     impute_function=impute)
```

```python
extraction_settings = ComprehensiveFCParameters()

X4 = extract_features(data4, column_id='id', column_sort='time',
                     default_fc_parameters=extraction_settings,
                     # we impute = remove all NaN features automatically
                     impute_function=impute)
```

```python
extraction_settings = ComprehensiveFCParameters()

X5 = extract_features(data5, column_id='id', column_sort='time',
                     default_fc_parameters=extraction_settings,
                     # we impute = remove all NaN features automatically
                     impute_function=impute)
```

```python
res = pd.concat([X1,X2,X3,X4,X5])
```

```python
res.describe()
```

```python
res.to_csv('features_dblm_1.csv', index=True, index_label=False)
```
